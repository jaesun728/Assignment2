```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

df = pd.read_csv('서울시 일별 평균 대기오염도 정보2.csv', index_col = '측정일시', encoding='cp949')
df
```

```python
df_avr = df.groupby('측정소명').mean()
df_avr
```

```python
main_df=df_avr[["미세먼지(㎍/㎥)","초미세먼지(㎍/㎥)"]].head(10)
main_df
```

```python
font_name = mpl.font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
mpl.rc('font', family=font_name)

ax = main_df.plot(kind='bar', title='날씨', figsize=(12, 4), legend=True, fontsize=12)
ax.set_xlabel('도시', fontsize=12)
ax.set_ylabel('미세먼지/초미세먼지', fontsize=12)
ax.legend(['미세먼지', '초미세먼지'], fontsize=12)
```
